This notebook plots the output of ``benchmark.py``

```python
import pandas as pd
from matplotlib import pyplot as plt
from benchmark import RANGE
```

```python
with open('out.dta') as fin:
    d = eval(fin.read())
```

```python
df = pd.DataFrame(d)
```

```python
df.index = RANGE
```

Note that below,
- ``gen`` refers to the C (generic) code path
- ``int`` refers to the Python (int-only) code path
- ``mix`` refers to a random array
- ``sep`` refers to an array which is the concatenation of two sorted sub-arrays

```python
df.plot(marker='.')
plt.grid()
plt.xlabel("N")
plt.savefig('abs.png')
```

```python
df.plot(marker='.')
plt.loglog()
plt.grid()
plt.xlabel("N")
plt.savefig('log.png')
```

```python
(df['gen_mix']/df['int_mix']).plot(marker='.')

plt.semilogx()
plt.grid()
plt.xlabel("N")
plt.savefig('log.png')
```
