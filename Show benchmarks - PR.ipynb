This notebook plots the output of ``benchmark_np_PR.py``

```python
import pandas as pd
from matplotlib import pyplot as plt
from benchmark_np_PR import RANGE
```

```python
data = {}

for label in 'PR', 'base':
    with open(f'out_np_{label}.dta') as fin:
        data[label] = pd.DataFrame(eval(fin.read()))
```

```python
df = pd.concat(data, axis=1)
```

```python
df
```

```python
df.index = RANGE
```

Note that below,
- ``mix`` refers to a random array
- ``sep`` refers to an array which is the concatenation of two sorted sub-arrays (important case for the application to COO sparse matrices in scipy)
- ``gen`` refers to the C (generic) code path
- ``int`` refers to the Python (int-only) code path with the default argsort ``kind`` (quicksort)
- ``sta`` refers to the Python (int-only) code path with ``kind=stable`` (which has linear cost in the ``sep`` case)


```python
df.drop([('base', 'sta_mix'), ('base', 'sta_sep')], axis=1).plot(marker='.')
plt.grid()
plt.xlabel("N")
plt.savefig('abs.png')
```

```python
df.drop([('base', 'sta_mix'), ('base', 'sta_sep')], axis=1).plot(marker='.')
plt.loglog()
plt.grid()
plt.xlabel("N")
plt.savefig('log.png')
```

```python
(df[('base', 'gen_mix')]/df[('PR', 'gen_mix')]).plot(marker='.')

plt.semilogx()
plt.grid()
plt.xlabel("N")
plt.ylim(0, 2.2)
plt.savefig('speedup.png')
```

```python
(df[('base', 'big_lex')]/df[('PR', 'big_lex')]).plot(marker='.')

plt.semilogx()
plt.grid()
plt.xlabel("N")
plt.ylim(0, 2.2)
plt.savefig('speedup_big.png')
```
